##**Interact with the operating system in Python**

**Summary:**
1.  `os.getcwd()`
2.  `os.listdir()`
3.  `os.system()`
4.  `os.listdir(path)`
5. `os.walk(path)`

```
1. Method to retrieve the current working directory path and store that value on the pwd variable.
2. Method to obtain the filenames and directories in the current working directory.
3. Allows us to execute a shell command.
4.  Returns a list with the contents of the directory passed as an argument.
5. Navigates all the directories in the provided path directory, and returns
three values: the path directory, the names of the subdirectories, and a list of filenames
in the current directory path.

```
